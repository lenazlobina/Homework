### шаг 2

```python
# с каждой страницы соберите ссылки на отдельные новости
# подсказка: родительский тег h2 (проверяйте по нему), нужный нам - a

all_links = []

# ваш код

for i in range(0,len(all_pages)):
    page = rq.get(all_pages[i])
    soup = BeautifulSoup(page.text, features="html.parser")
    for link in soup.find_all('a'):
        if link.parent.name == 'h2'and link.get('href') not in all_links:
            all_links.append(link.get('href')) 

all_links  
```

### шаг 4

```python
# пишем функцию по сбору информации
def GetNews(url0):
    page0 = rq.get(url0)
    soup0 = BeautifulSoup(page0.text, features="html.parser")
    author = soup0.find_all('meta', {'name' : 'author'})[0].attrs['content']
    date = soup0.find('time').text
    title = soup0.find('h1').text
    text_list = soup0.find_all('p')
    text = ' '.join([i.text.strip() for i in text_list]).replace('\xa0', ' ').replace('  ', '').replace('Получайте наши подборки первыми\n\n Согласен сполитикой конфиденциальности © 2018–2024 Системный Блокъ','')
    topics_list = soup0.find_all('a',{'rel':'category tag'})
    topics = ' '.join([i.text for i in topics_list])
    return url0, topics, author, date, title, text
```

```python
# собираем все новости!
# полезно использовать try / except
news = [] 

for link in all_links[:20]:
    try:
        new = GetNews(link)
        news.append(new)
    except:
        print(link)
```
